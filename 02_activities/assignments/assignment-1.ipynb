```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
 
# y-axis in bold
rc('font', weight='bold')
 
# Values of each group
bars1 = [12, 28, 1, 8, 22]
bars2 = [28, 7, 16, 4, 10]
bars3 = [25, 3, 23, 25, 17]
 
# Heights of bars1 + bars2
bars = np.add(bars1, bars2).tolist()
 
# The position of the bars on the x-axis
r = [0,1,2,3,4]
 
# Names of group and bar width
names = ['A','B','C','D','E']
barWidth = 1
 
# Create brown bars
plt.bar(r, bars1, color='#7f6d5f', edgecolor='white', width=barWidth)
# Create green bars (middle), on top of the first ones
plt.bar(r, bars2, bottom=bars1, color='#557f2d', edgecolor='white', width=barWidth)
# Create green bars (top)
plt.bar(r, bars3, bottom=bars, color='#2d7f5e', edgecolor='white', width=barWidth)
 
# Custom X axis
plt.xticks(r, names, fontweight='bold')
plt.xlabel("group")
 
# Show graphic
plt.show()
```

```python

np.random.seed(613) #set seed for reproducibility
#define data
x = np.arange(50)  #generate x values between 0 and 50
y1 = np.random.randint(0, 100,50)  #generate 50 random x values between 0 and 100
y2 = np.random.randint(0, 100,50)  #generate 50 random x values between 0 and 100
```

```python
#define fig size
fig, ax = plt.subplots(figsize=(5, 3))
#make plots
ax.plot(x,y1, color='red', label='person 1')
ax.plot(x,y2,color='blue', label='person 2')
#add legend
ax.legend(loc='lower right')
ax.legend(loc='lower right', frameon = True, #add frame around the legend 
          fontsize = 12, #change font size 
          ncol = 2, #specify number of columns 
          shadow = True )
```

```python
#define fig size
fig, ax = plt.subplots(figsize=(5, 3))
#make lineplot
ax.plot(x,y1, color='red', label='person 1')
ax.plot(x,y2,color='blue', label='person 2')
#add legend
ax.legend(loc='lower right', bbox_to_anchor = (1,1))
```

```python
np.random.seed(613) #set seed for reproducibility
#define data
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
#define plot layouts and size
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'], ['ax2', 'ax3']], 
                                   figsize=(7, 4),
                                   layout = 'constrained')
#create plots
someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)
#add labels
someaxes["ax1"].set_xlabel('scatter', fontsize=20)
someaxes["ax2"].set_xlabel('bar', fontsize=20)
someaxes["ax3"].set_xlabel('plot', fontsize=20)
plt.show()
```

```python
# load libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
#load data
tips = sns.load_dataset('tips')
print(tips)
```

```python
#set plot grid style
sns.set_style('darkgrid')
#create lineplot
sns.lineplot(data=tips, x = 'total_bill', y = 'tip')
```

```python
#cretae lineplot
fig = plt.subplots(figsize=(10, 3))
tipgraph = sns.lineplot(data=tips, x='total_bill', y='tip')
#add label
tipgraph.set(title = 'Trips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
#create lineplot with extra astethics
fig = plt.subplots(figsize=(10, 3))
tipgraph = sns.lineplot(data=tips, x='total_bill', y='tip',
                        color='hotpink',
                        linestyle = '--',
                        linewidth = 3,
                        marker = 'o',
                        markerfacecolor = 'indigo')
#add label
tipgraph.set(title = 'Trips vs Total Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
#create scatter plot
tipgraph = sns.scatterplot(data=tips, x='total_bill', y='tip', style='time', hue='day',
                           palette = ['purple','hotpink','deepskyblue','yellowgreen'])
#add label
tipgraph.set(title = 'Trips Vs Bill',
             xlabel = 'Total Bill ($)',
             ylabel = 'Tip Amount ($)')
```

```python
#see the different relationships within the data attributes
sns.pairplot(data=tips, hue='day')
```

```python
#Relplot--based on teh categories within the variable, color-code sex
daysplot = sns.relplot(data=tips, 
                       x='total_bill', 
                       y ='tip', 
                       hue = 'sex', 
                       col='day', 
                       kind='scatter',
                       col_wrap=2)
```

## Plotly Graphs

```python
#import libraries
import plotly.graph_objects as go
from wordcloud import WordCloud
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
#load data
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
```

```python
#Create the figure
graph = go.Figure()
#add data to the plot
graph.add_trace(go.Bar(x=x1, y=y1))
#add labels
graph.update_layout( title="Pirate Scores", 
                    xaxis_title="Pirates", 
                    yaxis_title="Score")
graph.show()
```

```python
#save the plot
graph.write_html("../pirategraph.html")
```

```python
#create an empty figure
graph = go.Figure()
#add data to the plot
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers', # we want points for a scatter plot 
                           marker=dict( size=15, # point size 
                                       color='hotpink', # point colour 
                                       opacity=1, # point transparency/alpha 
                                       line=dict(width=5, color='purple') # point outline 
                                       )))
#add labels
graph.update_layout(title='Interactive Pirate Plot', 
                    xaxis_title='Pirates', 
                    yaxis_title='Scores', 
                    width=500, 
                    height=500)
```

## WordCloud 

```python
#load data
df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                 on_bad_lines='skip')  #skips rows with bad columns
df
```

```python
# join all our text from each row from our quote column into a string
text = " ".join(each for each in df.quote)
# generate our wordcloud image
wordcloud = WordCloud(background_color="white",
                      colormap = 'inferno').generate(text)
# use matplotlib syntax to put our image in a figure
fig, ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, 
          interpolation='bilinear') # this line helps smooth our image
ax.axis("off")
plt.show()
```

## Venn Diagram

```python
#define data
A = set(["apple", "banana", "watermelon"])
B = set(["pumpkin", "blueberry", "apple", "key lime"])
```

```python
#plot the venn diagram
diagram = venn2_unweighted([A, B], 
                           set_labels = ('Fruits', 'Pies'), 
                           set_colors=("blue", "red"), 
                           alpha=0.5)
# Add labels and annotation
diagram.get_label_by_id("10").set_text("\n".join(A - B))
diagram.get_label_by_id("11").set_text("\n".join(A & B))
diagram.get_label_by_id("01").set_text("\n".join(B - A))
plt.show()
```
