```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load data
data = pd.read_csv('/Users/adaezeibik/Downloads/Child care centres - 4326.csv')
data
                   
```

```python
#set plot style
plt.style.use('ggplot')
```

```python
#count coulmn values uniquely
value_counts = data['IGSPACE'].value_counts()
```

```python
#define figure size
fig, ax = plt.subplots(figsize=(10, 6))

#create bar plot of data
value_counts.plot(kind='bar', color='red', label='IGSPACE means Infant Room Space')

ax.bar_label(ax.containers[0], label_type='edge')  # Automatically adds counts
#add labels and legend
ax.set_title('Number of Child care centers with available IGSPACE in Toronto Child care centers')
ax.set_ylabel('Number of Child care centers')
ax.set_xlabel('Number of IGSPACE')
plt.legend()
plt.show()
```
